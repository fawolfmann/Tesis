```python
import numpy as np
import tensorflow as tf
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
# Load TFLite model and allocate tensors.
interpreter = tf.lite.Interpreter(model_path='tflite_models/DenseNet_trigo_Xception_88.tflite')
interpreter.allocate_tensors()
```

```python
# Get input and output tensors.
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()
```

```python
IMAGE_SHAPE = (224 , 224, 3)
path  = './dataset/trigo/croped_test/sano/25_1.png'
image = mpimg.imread(path)
plt.imshow(image)
```

```python
img_arr = np.array(image, dtype=np.float32)
img_arr = np.resize(img_arr, IMAGE_SHAPE)
img_arr  =np.expand_dims(img_arr, 0)
np.amax(img_arr), img_arr.shape
```

```python
# Test model on random input data.
input_shape = input_details[0]['shape']
input_data = img_arr
interpreter.set_tensor(input_details[0]['index'], input_data)
```

```python
interpreter.invoke()
# The function `get_tensor()` returns a copy of the tensor data.
# Use `tensor()` in order to get a pointer to the tensor.
output_data = interpreter.get_tensor(output_details[0]['index'])
print(output_data)
```
