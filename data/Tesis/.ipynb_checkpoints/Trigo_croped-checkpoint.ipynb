```python
from __future__ import absolute_import, division, print_function, unicode_literals

import os
import numpy as np
import scipy 
import pathlib
import matplotlib.pyplot as plt
import tensorflow as tf
from PIL import Image
import IPython.display as display
import time

keras = tf.keras
print(tf.__version__)
```

## Get the data

```python
data_dir = './dataset/trigo/croped/'
data_dir = pathlib.Path(data_dir)
```

```python
image_count = len(list(data_dir.glob('*/*.png')))
image_count
```

```python
N = 0  # total files
for dirpath, dirnames, filenames in os.walk(data_dir):    
    dirpath = dirpath.split("/")[-1]
    if dirpath != ".ipynb_checkpoints":
        N_c = len(filenames)
        N += N_c
        print( dirpath+ ": -> " + str(N_c))
print( "Total Files " + str(N) )
```

```python
CLASS_NAMES = np.array([item.name for item in data_dir.glob('*') if item.name != ".ipynb_checkpoints"])
CLASS_NAMES
```

```python
# The 1./255 is to convert from uint8 to float32 in range [0,1].
image_generator = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255,
                                                                  validation_split=0.15,                            
                                                                  zoom_range=0.4,
                                                                  horizontal_flip=True,
                                                                  rotation_range=90,
                                                                 )
```

```python
BATCH_SIZE = 32
IMG_HEIGHT = 224
IMG_WIDTH = 224
# IMG_HEIGHT = 448
# IMG_WIDTH = 448
STEPS_PER_EPOCH = np.ceil((image_count*0.85)/BATCH_SIZE)
int(STEPS_PER_EPOCH)
```

```python
train_data_gen = image_generator.flow_from_directory(directory= str(data_dir),
                                                     batch_size= BATCH_SIZE,
                                                     shuffle= True,
                                                     target_size= (IMG_HEIGHT, IMG_WIDTH),
                                                     subset = "training",
                                                     # save_to_dir = '/tf/data/Tesis/dataset/augmented',
                                                     # save_prefix = 'trigo_',
                                                     classes = list(CLASS_NAMES))
```

```python
valid_data_gen = image_generator.flow_from_directory(directory= str(data_dir),
                                                     batch_size= BATCH_SIZE,
                                                     shuffle= True,
                                                     target_size= (IMG_HEIGHT, IMG_WIDTH),
                                                     subset = "validation",
                                                     classes = list(CLASS_NAMES))
```

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
sample_training_images, _ = next(train_data_gen)
```

```python
plotImages(sample_training_images[:5])
```

## Bring the models

## Xception

```python
base_model_x = tf.keras.models.load_model('models/plant_village_black_Xception_trained.h5')
```

```python
base_model_x.layers[0].traianble = False
```

```python
global_average_layer_x = tf.keras.layers.GlobalAveragePooling2D()
prediction_layer_x = tf.keras.layers.Dense(len(CLASS_NAMES), activation="softmax")

model_x = tf.keras.Sequential([
    base_model_x.layers[0],
    global_average_layer_x,
    prediction_layer_x
])
```

```python
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.3,
                              patience=2, min_lr=0.0001, verbose = 1, min_delta = 0.005)
early = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=4)
filepath_x = "models/plant_village_trigo_croped_Xception.{epoch:02d}-{val_accuracy:.2f}.h5"
checkpoint = tf.keras.callbacks.ModelCheckpoint(filepath = filepath_x, 
                                                          save_best_only=True , monitor='val_accuracy', mode='max')
```

```python
base_learning_rate = 0.03
model_x.compile(optimizer=tf.keras.optimizers.Adam(lr=base_learning_rate),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
total_epochs = 60
start_time = time.time()
history_x = model_x.fit_generator(train_data_gen,
                         epochs=total_epochs,
                         steps_per_epoch=train_data_gen.__len__(),
                         validation_data=valid_data_gen,
                         validation_steps= valid_data_gen.__len__(),
                         callbacks = [reduce_lr , checkpoint]
                        )
duration = time.time() - start_time
print('took: ' + str(duration/60))
```

```python
for i, layer in enumerate(model_x.layers[0].layers):
   print(i, layer.name)
```

```python
for i in range(len((model_x.layers[0].layers)))[115:]:
    model_x.layers[0].layers[i].trainable = True
```

```python
for i, layer in enumerate(model_x.layers[0].layers):
    if str(layer)[-2] == 'bn':
        model_x.layers[0].layers[i].trainable = True
    elif 'batch_normalization' == str(layer):
        model_x.layers[0].layers[i].trainable = True
```

```python
base_learning_rate = 0.03
model_x.compile(optimizer=tf.keras.optimizers.Adam(lr=base_learning_rate),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model_x.summary()
```

```python
total_epochs = 30
start_time = time.time()
history_x = model_x.fit_generator(train_data_gen,
                         epochs=total_epochs,
                         steps_per_epoch=train_data_gen.__len__(),
                         validation_data=valid_data_gen,
                         validation_steps= valid_data_gen.__len__(),
                         callbacks = [reduce_lr , checkpoint]
                        )
duration = time.time() - start_time
print('took: ' + str(duration/60))
```

```python
base_learning_rate = 0.003
model_x.compile(optimizer=tf.keras.optimizers.Adam(lr=base_learning_rate),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
initial_epochs = 30
final_epochs = 30
total_epochs = initial_epochs + final_epochs
start_time = time.time()
history_x = model_x.fit_generator(train_data_gen,
                         epochs=total_epochs,
                         initial_epoch = initial_epochs,
                         steps_per_epoch=train_data_gen.__len__(),
                         validation_data=valid_data_gen,
                         validation_steps= valid_data_gen.__len__(),
                         callbacks = [reduce_lr , checkpoint]
                        )
duration = time.time() - start_time
print('took: ' + str(duration/60))
```

```python
for i in range(len((model_x.layers[0].layers)))[85:]:
    model_x.layers[0].layers[i].trainable = True
```

```python
base_learning_rate = 0.003
model_x.compile(optimizer=tf.keras.optimizers.Adam(lr=base_learning_rate),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
initial_epochs = 60
final_epochs = 30
total_epochs = initial_epochs + final_epochs
start_time = time.time()
history_x_fine = model_x.fit_generator(train_data_gen,
                         epochs=total_epochs,
                         initial_epoch = initial_epochs,
                         steps_per_epoch=train_data_gen.__len__(),
                         validation_data=valid_data_gen,
                         validation_steps= valid_data_gen.__len__(),
                         callbacks = [reduce_lr , checkpoint]
                        )
duration = time.time() - start_time
print('took: ' + str(duration/60))
```

```python
base_learning_rate = 0.003
model_r.compile(optimizer=tf.keras.optimizers.Adam(lr=base_learning_rate),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model_r.summary()
```

```python
total_epochs = 60
start_time = time.time()
history_r = model_r.fit_generator(train_data_gen,
                         epochs=total_epochs,
                         steps_per_epoch=train_data_gen.__len__(),
                         validation_data=valid_data_gen,
                         validation_steps= valid_data_gen.__len__(),
                         callbacks = [reduce_lr , checkpoint]
                        )
duration = time.time() - start_time
print('took: ' + str(duration/60))
```
